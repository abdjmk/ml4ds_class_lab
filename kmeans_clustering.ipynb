```python
from sklearn import datasets, cluster
import matplotlib. pyplot as plt
import numpy as np
blob_values, blob_labels = datasets.make_blobs(n_samples = 300, centers= 3)
print(blob_values)

print(blob_labels)

print("Orange = 0, Green =1, & Blue =2")

for index in range(len(blob_values)):
	if blob_labels[index]==0:
		plt.scatter(blob_values[index][0],blob_values[index][1],c="orange")
	elif blob_labels[index]==1:
		plt.scatter(blob_values[index][0],blob_values[index][1],c="green")
	else:
		plt.scatter(blob_values[index][0],blob_values[index][1],c="blue")
plt.show()
```

```python
our_kmeans_cluster = cluster.KMeans(n_clusters=3) 
our_kmeans_cluster.fit(blob_values)
print(our_kmeans_cluster)
```

```python
center1,center2, center3 =our_kmeans_cluster.cluster_centers_
print(center1)
print(center2)
print(center3)
```

```python
### 800 sample ###


from sklearn import datasets, cluster
import matplotlib. pyplot as plt
import numpy as np
blob_values, blob_labels = datasets.make_blobs(n_samples = 800, centers= 5)
print(blob_values)

print(blob_labels)

print("Orange = 0, Green =1, Blue =2, Violet =3, & Indigo=4")

for index in range(len(blob_values)):
	if blob_labels[index]==0:
		plt.scatter(blob_values[index][0],blob_values[index][1],c="orange")
	elif blob_labels[index]==1:
		plt.scatter(blob_values[index][0],blob_values[index][1],c="green")
	elif blob_labels[index]==2:
		plt.scatter(blob_values[index][0],blob_values[index][1],c="blue")
	elif blob_labels[index]==3:
		plt.scatter(blob_values[index][0],blob_values[index][1],c="violet")
	else:
		plt.scatter(blob_values[index][0],blob_values[index][1],c="indigo")
plt.show()
```
