```python
import numpy
import matplotlib.pyplot as plt
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]
mymodel = numpy.poly1d(numpy.polyfit(x, y, 3)) 
myline = numpy.linspace(1, 22, 100)
plt.scatter(x, y)
plt.plot(myline, mymodel(myline))

#Display the diagram:

# plt.vlines(20 , ymax=mymodel(20), ymin = 50, colors='g')

# plt.hlines(mymodel(20), xmax=20, xmin = 0, colors = 'g')

# plt.show()

plt.vlines(5, ymax=mymodel(5), ymin = 50, colors='g')

plt.hlines(mymodel(5), xmax=5, xmin = 0, colors = 'g')

plt.show()
```
